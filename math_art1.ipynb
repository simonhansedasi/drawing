```python
import numpy as np
import matplotlib.pyplot as plt

def math_art(x, y):
    return (np.sin(x) - np.cos(y)) * (np.cos(x) + np.sin(y))

x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.linspace(-2*np.pi, 2*np.pi, 300)
X, Y = np.meshgrid(x, y)

Z = math_art(X, Y)

plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def math_art(x, y, phase):
    return np.sin(x + phase) + np.cos(y + phase) * (np.sin(-x + phase) + np.cos(-y + phase))
x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.linspace(-2*np.pi, 2*np.pi, 300)
X, Y = np.meshgrid(x, y)

frames = []
phases = np.linspace(0, 2*np.pi, 60)

for phase in phases:
    Z = math_art(X, Y, phase)
    
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.contourf(X, Y, Z, cmap='viridis')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.gca().set_aspect('equal')
    
    # Save the current frame with tight layout
    fig.tight_layout(pad=0)
    fig.canvas.draw()
    
    # Save the figure to a file-like buffer
    buf = fig.canvas.tostring_rgb()
    ncols, nrows = fig.canvas.get_width_height()
    image = Image.frombytes('RGB', (ncols, nrows), buf)
    
    # Crop the image to remove white borders
    bbox = image.getbbox()
    image = image.crop(bbox)
    
    frames.append(image)
    plt.close(fig)

# Save the frames as a GIF
gif_filename = 'rotating_math_art.gif'
frames[0].save(gif_filename, save_all=True, append_images=frames[1:], duration=100, loop=0)

# Display the GIF
from IPython.display import Image as IPImage
IPImage(filename=gif_filename)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def math_art(x, y, phase):
    return np.sin(x + phase) + np.sin(y + phase) * (np.cos(-x + phase) - np.cos(-y + phase))
x = np.linspace(-2*np.pi, 2*np.pi, 300)
y = np.linspace(-2*np.pi, 2*np.pi, 300)
X, Y = np.meshgrid(x, y)

frames = []
phases = np.linspace(0, 2*np.pi, 60)

for phase in phases:
    Z = math_art(X, Y, phase)
    
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.contourf(X, Y, Z,cmap = 'coolwarm')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.gca().set_aspect('equal')
    
    # Save the current frame with tight layout
    fig.tight_layout(pad=0)
    fig.canvas.draw()
    
    # Save the figure to a file-like buffer
    buf = fig.canvas.tostring_rgb()
    ncols, nrows = fig.canvas.get_width_height()
    image = Image.frombytes('RGB', (ncols, nrows), buf)
    
    # Crop the image to remove white borders
    bbox = image.getbbox()
    image = image.crop(bbox)
    
    frames.append(image)
    plt.close(fig)

# Save the frames as a GIF
gif_filename = 'rotating_math_art.gif'
frames[0].save(gif_filename, save_all=True, append_images=frames[1:], duration=100, loop=0)

# Display the GIF
from IPython.display import Image as IPImage
IPImage(filename=gif_filename)
```
